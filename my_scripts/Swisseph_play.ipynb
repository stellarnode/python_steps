```python
import swisseph as swe
from datetime import datetime

swe.set_ephe_path('/Users/stellarnode/Development/swisseph/ephe')

swe.set_sid_mode(swe.SIDM_LAHIRI)

now = datetime.now()
print("normal date ", now)
now = swe.julday(now.year, now.month, now.day, now.hour + now.minute / 60, cal=swe.GREG_CAL)
print("jul date ", now)
print(swe.FLG_SIDEREAL)
pack = swe.calc_ut(now, swe.MOON, swe.FLG_SIDEREAL)
print("pack ", pack)

# res, flg = swe.calc_ut(now, swe.MOON)
nak, pada = swe._long2nakshatra(pack[0])
print('Nakshatra = %d, pada = %d' % (nak + 1, pada + 1))
nak_name = swe._get_nakshatra_name(nak)
print(nak_name)

print(str(math.floor(pack[0])) + " deg " + str((pack[0] - math.floor(pack[0])) * 60) + " min")
```
