```python
import os
```

```python
files_path = '/Volumes/Magic Power/(V) Videos & Movies/TV-SERIES/Mahakali (2017)/Season 01'
```

```python
file_list = os.listdir(files_path)
file_list
```

```python
def add_zeroes(number):
    out = str(number)
    prefix = ''
    if number < 10:
        prefix = '00'
    elif number < 100:
        prefix = '0'
    else:
        prefix = ''
    return prefix + out
```

```python
for count, filename in enumerate(file_list):
    full_path = os.path.join(files_path, filename)
    print(full_path)
    if os.path.isfile(full_path):
        name = full_path.split('.')
        dest = name[0] + '_s01e' + add_zeroes(count + 1) + '.' + name[2]
        print(dest)
        os.rename(full_path, dest)
```
